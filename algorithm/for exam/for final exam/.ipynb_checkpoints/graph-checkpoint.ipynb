```python
# BFS and DFS

BFS(r):
    for v in V:
        visited[v] = No
    visited[r] = Yes
    
    Enqueue(Q, r)
    
    while Q != 0:
        u = Dequeue(Q)
        
        for v in L(u):
            if visited[v] == No:
                visited[v] = Yes
                Enqueue(Q, v)
                
DFS(u):
    visited[u] = Yes
    for v in L(u):
        if visited[v] == No:
            DFS(v)
            
DFS():
    for v in V:
        visited[v] = No
    
    for v in V:
        if visited[v] == No:
            aDFS(v)

aDFS(v):
    visited[v] = Yes
    for v in L(v):
        if visited[v] == No:
            aDFS(v)
    
```

```python
# 최소신장트리
prim(r):
    S = {}
    for v in V:
        d[v] = INF
    d[r] = 0
    
    while V != S:
        u = extractMin(V-S, d)
        S U {u}
        for v in L(u):
            if v in V-S and w[u][v] < d[v]:
                d[v] = w[u][v]
                prev[v] = u

kruskal():
    S = {}
    for v in V:
        make_set(v)
    # edge집합을 정렬한다.
    
    for u, v in E:
        if find_set(u) != find_set(v):
            S = S U {u, v}
            Union(find_set(u), find_set(v))
    return S
```

```python
# 최단경로
digistra(r):
    S = {}
    for v in V:
        d[v] = INF
    d[r] = 0
    
    while V != S:
        u = extractMin(V-S, d)
        S U {u}
        for v in L(u):
            cost = d[u] + w[u][v]
            if v in V-S and cost < d[v]:
                d[v] = cost
                prev[v] = u

bell_ford(r):
    for v in V:
        d[v] = INF
    d[r] = 0
    
    for i in 1 to V.length-1:
        for u, v in E:
            cost = d[u] + w[u][v]
            if cost < d[v]:
                d[v] = cost
                prev[v] = u
    
    for u, v in E:
        cost = d[u] + w[u][v]
        if cost < d[v]:
            return "해없음"
    
floyd():
    n = V.length
    for i in 1 to n:
        for j in 1 to n:
            dp[i][j] = w[i][j]
    
    for k in 1 to n:
        for i in 1 to n:
            for j in 1 to n:
                dp[i][j] = min(dp[i][j], dp[i][k] + dp[k][j])
                
DAG(r):
    for v in V:
        d[v] = INF
    d[r] = 0
    #그래프를 위상정렬한다.
    for v in V:  #V는 위상정렬된 순서
        for u in L(v):
            cost = d[v] + w[v][u]
            if cost < d[u]:
                d[u] = cost
```
